#Introducción.

TensorFlow es un framework de Deep Learning desarrollado por Google, a lo largo de su desarrollo a sufrido diversas modificaciones, esto es evidente si comparamos TensorFlow 1.x con la version 2.x, la version 2.x está mas enfocada a utilizar APIs de alto nivel, haciendo las implementaciones de modelos de aprendizaje generales mas sencillas.

A continuacion se mostrará un ejemplo de una implementacion de TensorFlow v1.X

Código usado para Tensorflow 1.x


```python
import tensorflow as tf
message = tf.constant('Welcome to the exciting world of Deep Neural Networks!')
with tf.Session() as sess:
  print(sess.run(message).decode())
```

Correr un programa de Tensorflow 1.x en 2.x

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
in_a = tf.placeholder(dtype=tf.float32, shape=(2))
def model(x):
  with tf.variable_scope("matmul"):
    W = tf.get_variable("W", initializer=tf.ones(shape=(2,2)))
    b = tf.get_variable("b", initializer=tf.zeros(shape=(2)))
    return x * W + b

out_a = model(in_a)

with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())
  outs = sess.run([out_a],
      feed_dict={in_a: [1, 0]})
  writer = tf.summary.FileWriter("./logs/example", sess.graph)
```

##Conclusion.

Ambas versiones tienen sus ventajas, la version 2 permite implementar modelos con mayor facilidad, por otro lado la version 1 permite controlar detalles muy especificos de los modelos de aprendizaje, afortunadamente para nosotros, entrando a los especificos TFv1.x, trabaja solamente con grafos de computacion estatica y por otra parte TFv2.x, trabaja con grafos de computacion dinamica y estática. Ademas TFv2.x cuenta con Keras integrado como API de alto nivel.
